### P1. Write a Python code to output the following graph. (or a similar one)
```
The productivity values for the 3 groups are the following: 

g1 = [80, 60, 80, 67, 34]
g2 = [45, 67, 23, 89, 10]
g3 = [30, 90, 50, 20, 90]
```

````python
import matplotlib.pyplot as plt
import numpy as np

groups = np.array([[80, 60, 80, 67, 34],[45, 67, 23, 89, 10],[30, 90, 50, 20, 90]])
index = np.arange(5)
bar_width = 0.2
ticks = ['M', 'T', 'W', 'Th', 'F']

plt.bar(index, groups[0], bar_width, label="Group1")                #tick_label=ticks
plt.bar(index+bar_width, groups[1], bar_width, label="Group2")
plt.bar(index+2*bar_width, groups[2], bar_width, label="Group3")    #, align='center'

plt.xticks(index + bar_width, ticks)

plt.legend()
plt.xlabel("Groups")
plt.ylabel("Productivity")
plt.title("Productivity during the week")

plt.show()
````

# Version 2 via pandas

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


prod = pd.DataFrame(np.array([[80, 60, 80, 67, 34],[45, 67, 23, 89, 10],[30, 90, 50, 20, 90]]),index=['Group1','Group2','Group3'],columns=['M','T', 'W', 'Th', 'F'])
print(prod)

ax = prod.T.plot(kind='bar', title="Productivity during the week", rot=1)
ax.set(xlabel="Groups", ylabel="Productivity")

````

### P2. Write a Python code to output the following graph. (or a similar one)
```
Draw 25 dots. Generate random x and y coordinates (integers from 1 to 30), random colors and random sizes (integers from 20 to 500) for the dots.
```

```
Նկարեք 25 կետեր. Կետերի համար գեներացրեք պատահական x և y կոորդինատներ (1-ից 30 միջակայքում int տիպի թվեր), պատահական գույներ ու պատահական չափսեր (20-ից 500 միջակայքում int տիպի թվեր)։
```

````python
import matplotlib.pyplot as plt
import numpy as np

# np.random.seed(90)
plt.scatter(np.random.randint(1,30,25), np.random.randint(1,30,25), c=np.random.randint(20,500,25), s=np.random.randint(20,500,25))
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0,30)
plt.title("Colorful dots")
plt.show()
````

# Version 2

````python
import matplotlib.pyplot as plt
import numpy as np

# np.random.seed(90)
col_lst = ['b', 'g', 'r', 'c','m','y','k']
col = np.random.choice(col_lst,25)
plt.scatter(np.random.randint(1,30,25), np.random.randint(1,30,25), c=col, s=np.random.randint(20,500,25))
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0,30)
plt.title("Colorful dots")
plt.show()
````
